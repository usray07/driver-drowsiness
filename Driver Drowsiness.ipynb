```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report

```

```python
import cv2
import glob
import numpy as np

X = []
files = glob.glob("train\Open_Eyes\*.png")
for myFile in files:
    image = cv2.imread (myFile,0)
    image = cv2.resize(image, (80, 80))
    X.append(image)
```

```python
files = glob.glob ("train\Closed_Eyes\*.png")
for myFile in files:
    image = cv2.imread (myFile,0)
    image = cv2.resize(image, (80, 80))
    X.append(image)
```

```python
y_data = np.array(y)
```

```python
#test data has 1000 images
#training using 3000 images
X_train,X_test,y_train,y_test = train_test_split(X_data,y_data,random_state=42,test_size=0.25)
```

```python
model = LogisticRegression(solver='liblinear', random_state=0)
```

```python
#.reshape( X_train3000, 6400)
```

```python
model.fit(X_train.reshape(X_train.shape[0], 80*80),y_train)
```

```python
pred = model.predict(X_test.reshape(X_test.shape[0], 80*80))
accuracy = accuracy_score(y_test, pred)
accuracy
```
